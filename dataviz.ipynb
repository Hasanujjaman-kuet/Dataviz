```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Line

```python
x = np.linspace(1,10, 100)
y1 = np.cos(x)
y2 = np.sin(x)
```

```python
plt.figure(1, figsize=(10, 6))
plt.plot(x, y1, label='COS Wave', linewidth=3, color='lightgreen', marker='^')
plt.plot(x, y2, label='SIN Wave',linewidth=5, color='lightblue', marker='o')
plt.title("X vs cos(X)")
plt.xlabel("x-axis")
plt.ylabel("cos(x)")
plt.legend()
plt.grid(True)
plt.xticks(np.arange(0, 10, 0.5))
plt.yticks(np.arange(-1, 1, 0.1))
plt.xlim(1, 9.0)
plt.ylim(-1, 0.8)
plt.show()
```

# Histogram

```python
var = np.random.randn(100)
```

```python
plt.figure(1, figsize=(10, 6))
plt.hist(var, bins=15, label='Var Distribution', linewidth=3, color='lightgreen')
plt.hist(y2, bins=15, label='Var Distribution', linewidth=3, color='orange')
plt.title("Var Distribution")
plt.xlabel("range")
plt.ylabel("frequescies")
plt.legend()
plt.grid(True)
plt.show()
```

# Bar Plot

```python
lang = ['Python', 'Java', 'C++', 'R']
ratings = [200, 180, 190, 167]
plt.figure(1, figsize=(10, 6))
plt.bar(lang, ratings, label='Language Rating', linewidth=3, color='lightgreen')
plt.title("Language Rating")
plt.xlabel("Languages")
plt.ylabel("Ratings")
plt.legend()
plt.grid(True)
plt.show()
```

```python
lang = ['Python', 'Java', 'C++', 'R']
ratings = [200, 180, 190, 167]
plt.figure(1, figsize=(10, 6))
plt.barh(lang, ratings, label='Language Rating', linewidth=3, color='lightgreen')
plt.title("Language Rating")
plt.ylabel("Languages")
plt.xlabel("Ratings")
plt.legend()
plt.grid(True)
plt.show()
```

# Pie Chart

```python
lang = ['Python', 'Java', 'C++', 'R']
ratings = [200, 140, 170, 117]
plt.figure(1, figsize=(10, 6))
plt.pie(ratings, labels=lang, 
        autopct='%.3f', explode=(0, 0, 0, 0.07), shadow=(0, 0, 0, 0.5))
plt.show()
```

# Scatter Plot

```python
x = np.random.randn(200)
y = np.random.randn(200)
plt.figure(1, figsize=(10, 6))
plt.scatter(x, y, label='Data Points', color='lightgreen')
plt.scatter(x, y*x/2, label='x*y/2 Points', color='orange')
plt.title("Data Points")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

# 3D Plots

```python
axes = plt.axes(projection='3d')
z = y*x/2
axes.scatter3D(x,y,z, label='Data')
axes.set_xlabel('x-axis')
axes.set_ylabel('y-axis')
axes.set_zlabel('z-axis')
axes.legend()
plt.savefig('plot3d.jpg')
```

# Multiple Subplots

```python
fig, axes = plt.subplots(nrows=2, ncols=3)
# fig.set(figsize=(20,8))
axes[0,0].plot(y1, y2)
axes[0,1].scatter(x, z)
axes[0,2].pie(ratings, labels=lang)
axes[1,0].hist(z)
axes[1,1].bar(lang, ratings)
axes[1,2].barh(lang, ratings)
plt.tight_layout()
```
